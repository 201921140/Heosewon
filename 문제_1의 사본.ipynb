##주어진 과일이름(사과, 포도, 복숭아, 참외, 수박, 바나나)에 대해 fruits란 리스트를 만들고 각각을 format을 활용해 print 하시오

```python
fruits=["사과", "포도", "복숭아", "참외", "수박", "바나나"]
print(("{},{},{},{},{},{}").format(fruits[0],fruits[1],fruits[2],fruits[3],fruits[4],fruits[5]))
```

##fruits에서 복숭아와 바나나 값을 제거하고 멜론과 파인애플을 추가하시오

```python
fruits.pop(2)
fruits.pop(4)
fruits.append("멜론")
fruits.append("파인애플")
print(fruits)
```
